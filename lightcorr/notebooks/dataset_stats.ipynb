```python
# get min max
min = np.inf
max = 0

for data in deepcorr_dataset:
    # Iterate through the dictionary to print the sizes of the lists under the arrow keys
    for key, tuples in data.items():
        #print(f"Key: {key}")
        for i, dict_ in enumerate(tuples):
            size_left = len(dict_["<-"])
            size_right = len(dict_["->"])
            #print(f" '<-': {size_left}, \n'->': {size_right}")
            if size_left < min:
                min = size_left
            if size_left > max:
                max = size_left
            if size_right < min:
                min = size_right
            if size_right > max:
                max = size_right

print(f"Min: {min}, Max: {max}")
```
