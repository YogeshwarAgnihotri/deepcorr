```python
three_hundert = []

for data in deepcorr_dataset_300:
    # Iterate through the dictionary to print the sizes of the lists under the arrow keys
    for key, tuples in data.items():
        #print(f"Key: {key}")
        for i, dict_ in enumerate(tuples):
            three_hundert.append(dict_["<-"])
            three_hundert.append(dict_["->"])
```

```python
double_count = 0

for data in deepcorr_dataset_400:
    # Iterate through the dictionary to print the sizes of the lists under the arrow keys
    for key, tuples in data.items():
        #print(f"Key: {key}")
        for i, dict_ in enumerate(tuples):
            # check if the data, if its in the list print both
            if dict_["<-"] in three_hundert:
                print(f"Found in 300: {dict_['<-']}")
                print(f"Found in 400: {dict_['<-']}")
                double_count += 1
```
